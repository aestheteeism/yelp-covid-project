## Flatten dataset

```python
import numpy as np
import json

def generate_covid_highlights_df(df):
    highlight_lists = list(df['highlights'])
    highlights = set()
    for hl in highlight_lists:
        if hl == False:
            continue
        for highlight in json.loads(hl):
            highlights.add(highlight['identifier'])

    values_for_highlight = dict()
    for highlight in highlights:
        values_for_highlight[highlight] = np.zeros((len(df)), dtype=bool)

    for i in range(len(highlight_lists)):
        if highlight_lists[i] == False:
            continue
        for highlight in json.loads(highlight_lists[i]):
            values_for_highlight[highlight['identifier']][i] = 1    

    for highlight in highlights:
        df[highlight] = values_for_highlight[highlight]  
    return df, highlights
```

```python
df, highlights = generate_covid_highlights_df(yelp_open)
highlights
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import SVMSMOTE 
from imblearn.under_sampling import RandomUnderSampler

def balance(X, y): # Oversample Yes 
    countY = y.value_counts().sort_values(ascending=False)
    if (countY[0] - countY[1]) / countY[1] <= 0.5:
        over = SMOTE(sampling_strategy='minority', random_state=42)
        X_res, y_res = over.fit_resample(X, y)
    else: 
        ratio = (5 * countY[1]) / countY[0]
        over = SVMSMOTE(sampling_strategy=ratio, random_state=42)
        #ada = ADASYN(sampling_strategy=ratio, random_state=42)
        X_res, y_res = over.fit_resample(X, y)
        under = RandomUnderSampler(sampling_strategy=0.2, random_state=42)
        X_res, y_res = under.fit_resample(X_res, y_res)
    return X_res, y_res
```

## Cross-fold validation

```python
target = 'is_closed_2021'

from sklearn.linear_model  import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold
from pandas import DataFrame
from sklearn.metrics import recall_score
from sklearn.preprocessing import StandardScaler

recalls = [] #some metric computed on each fold...

foldCount = 0
kf = StratifiedKFold(n_splits=10) #10 cross-fold validation is the gold standard
for train, test in kf.split(df.drop(target,axis=1), df[target]): #for each fold
    print("Processing fold %s" % foldCount)
    train_fold = df.iloc[train, :] #it’s important to separate training from testing data!
    test_fold = df.iloc[test, :]
    
    scaler = StandardScaler()
    
    # Balance 
    X_res, y_res = balance(train_fold.drop(target,axis=1), train_fold[target]) #pre-process training
    # Scale 
    X_res = scaler.fit_transform(X_res)
    logreg = LogisticRegression()
    logreg.fit(X_res, y_res) #fit on training
    
    X_test = test_fold.drop(target, axis=1)
    
    normalized_X_test = pd.DataFrame(scaler.transform(X_test), columns = X_test.columns)
    
    pred = logreg.predict(normalized_X_test)#predict using testing data
    
    print(recall_score(test_fold[target], pred))
    
    recalls.append(recall_score(test_fold[target], pred)) #measure quality
    foldCount += 1
    
print(DataFrame(recalls).mean()) #provide the mean metric. Optionally, variance too.
```

```python
from sklearn.metrics import recall_score
from sklearn.model_selection import ParameterGrid, StratifiedKFold
from sklearn.tree import DecisionTreeClassifier

parameter_grid = {
    'max_depth': [2,3,4,5],
    'min_samples_split': [2,3,5,6]
}


def nested_cv(X, y, inner_cv, outer_cv, Classifier, parameter_grid):
    outer_scores = []
    for training_samples, test_samples in outer_cv.split(X, y):
        best_params = []
        best_score = -np.inf
        for parameters in parameter_grid:
            cv_scores = []
            for inner_train, inner_test in inner_cv.split(X.iloc(axis=0)[training_samples], y.iloc(axis=0)[training_samples]):
                clf = Classifier(**parameters) 
                X_res, y_res = balance(X.iloc(axis=0)[inner_train], y.iloc(axis=0)[inner_train])
                clf.fit(X_res, y_res)
                score = clf.score(X.iloc(axis=0)[inner_test], y.iloc(axis=0)[inner_test])
                cv_scores.append(score)
            mean_score = np.mean(cv_scores)
            if mean_score > best_score:
                best_score = mean_score
                best_params = parameters
        clf = Classifier(**best_params)
        clf.fit(X.iloc(axis=0)[training_samples], y.iloc(axis=0)[training_samples])
        pred = clf.predict(X.iloc(axis=0)[test_samples])
        outer_scores.append(clf.score(X.iloc(axis=0)[test_samples], y.iloc(axis=0)[test_samples]))
    return np.array(outer_scores), best_params
```

```python
scores, best_params = nested_cv(df.drop(target,axis=1), df[target], StratifiedKFold(10),
                   StratifiedKFold(10), DecisionTreeClassifier, 
                   ParameterGrid(parameter_grid))
```
