## Flatten dataset

```python
import numpy as np
import json

def flatten_highlights(df):
    highlight_lists = list(df['highlights'])
    highlights = set()
    for hl in highlight_lists:
        if hl == False:
            continue
        for highlight in json.loads(hl):
            highlights.add(highlight['identifier'])

    highlight_dict = dict()
    for highlight in highlights:
        highlight_dict[highlight] = np.zeros((len(df)), dtype=bool)

    for i in range(len(highlight_lists)):
        if highlight_lists[i] == False:
            continue
        for highlight in json.loads(highlight_lists[i]):
            highlight_dict[highlight['identifier']][i] = 1    

    for highlight in highlights:
        df[highlight] = highlight_dict[highlight]  
    return df, highlights
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from imblearn.over_sampling import SVMSMOTE 
from imblearn.under_sampling import RandomUnderSampler

def balance(X, y): 
    countY = y.value_counts().sort_values(ascending=False)
    over = SVMSMOTE(sampling_strategy=(1.5 * countY[1]) / countY[0], random_state=42)
    X_res, y_res = over.fit_resample(X, y)
    under = RandomUnderSampler(sampling_strategy=0.5, random_state=42)
    X_res, y_res = under.fit_resample(X_res, y_res)
    return X_res, y_res
```

# AutoML

```python
import h2o
from h2o.automl import H2OAutoML
h2o.init(max_mem_size='16G')
```

```python
# Load data into H2O
df = pd.read_csv('new_data.csv')
df.head()
```

```python
print(f'Size of data set: {df.shape[0]} rows and {df.shape[1]} columns')
```

## Data processing with SMOTE

```python
df.replace({False: 0, True: 1}, inplace=True)
```

```python
ax = sns.countplot(x="is_closed_2021", data=df)
ax.set_title('Class distribution\n(0: Still open | 1: Permanently closed)')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
target = 'is_closed_2021'
X_train, X_test, y_train, y_test = train_test_split(df.drop(target,axis=1), df[target], test_size=0.2, random_state=42, stratify=df[target])
```

```python
X_train
```

```python
X_train.isna().sum()
```

```python
y_train.value_counts()
```

```python
# X_train_res, y_train_res = balance(X_train, y_train)
X_train_res, y_train_res = X_train, y_train
y_train_res.value_counts()
```

```python
df2 = X_train_res
df2['is_closed_2021'] = y_train_res
df2
```

```python
df2.dtypes
```

```python
train = h2o.H2OFrame(df2)
```

```python
df_test = X_test
df_test['is_closed_2021'] = y_test
test = h2o.H2OFrame(df_test)
```

```python
y = "is_closed_2021" 
x = train.columns 
x.remove(y) 

# For binary classification, response should be a factor
train[y] = train[y].asfactor()
test[y] = test[y].asfactor()
```

```python
aml = H2OAutoML(max_runtime_secs=120, seed=1, balance_classes=True)
aml.train(x=x,y=y, training_frame=train)
```

```python
lb = aml.leaderboard
lb.head()
```

```python
preds = aml.leader.predict(test)
preds
```

```python
df_output = test.cbind(preds)
df_output.head()
```

```python
df_output_pd = h2o.as_list(df_output)
df_output_pd
```

```python
from sklearn.metrics import classification_report

print(classification_report(df_output_pd['is_closed_2021'], df_output_pd['predict']))
```
