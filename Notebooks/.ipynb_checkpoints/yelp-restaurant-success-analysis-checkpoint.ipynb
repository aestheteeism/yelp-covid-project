```python
aDict = {"TRUE": True, "FALSE": False}

# Non-Exhaustive Mapping
cols = ['highlights', 'delivery or takeout', 'Grubhub enabled',
       'Call To Action enabled', 'Request a Quote Enabled', 'Covid Banner',
       'Temporary Closed Until', 'Virtual Services Offered']

def toBoolean(df, cols, di):
    for col in cols:
        df[col] = df[col].map(di).fillna(df[col])
    return df 

yelp_merged = toBoolean(yelp_merged, cols, aDict)
yelp_merged
```

## Flatten dataset

```python
import numpy as np
def generate_covid_highlights_df(df):
    highlight_lists = list(df['highlights'])
    highlights = set()
    for hl in highlight_lists:
        if hl == False:
            continue
        for highlight in json.loads(hl):
            highlights.add(highlight['identifier'])

    values_for_highlight = dict()
    for highlight in highlights:
        values_for_highlight[highlight] = np.zeros((len(df)), dtype=bool)

    for i in range(len(highlight_lists)):
        if highlight_lists[i] == False:
            continue
        for highlight in json.loads(highlight_lists[i]):
            values_for_highlight[highlight['identifier']][i] = 1    

    for highlight in highlights:
        df[highlight] = values_for_highlight[highlight]  
    return df, highlights
```

```python
df, highlights = generate_covid_highlights_df(yelp_merged)
highlights
```

```python
from sklearn.model_selection import train_test_split
target = 'Temporary Closed Until'
X_train, X_test, y_train, y_test = train_test_split(df2.drop(target,axis=1), df2[target], test_size=0.33, random_state=42, stratify=df2[target])
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

def balance(X, y): # Oversample Yes 
    countY = y.value_counts().sort_values(ascending=False)
    if (countY[0] - countY[1]) / countY[1] <= 0.5:
        over = SMOTE(sampling_strategy='minority', random_state=42)
        X_res, y_res = over.fit_resample(X, y)
    else: 
        ratio = (1.5 * countY[1]) / countY[0]
        over = SMOTE(sampling_strategy=ratio, random_state=42)
        X_res, y_res = over.fit_resample(X, y)
        under = RandomUnderSampler(sampling_strategy ='majority', random_state=42)
        X_res, y_res = under.fit_resample(X_res, y_res)
    return X_res, y_res
```

```python
from collections import Counter

# transform the dataset
X_train, y_train = balance(X_train, y_train)
# summarize the new class distribution
y_train.value_counts()
```

```python
from sklearn.linear_model  import LogisticRegression

clf = LogisticRegression(C = 50)

clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report

clf_predicted = clf.predict(X_test)
confusion = confusion_matrix(y_test, clf_predicted)
confusion
```
