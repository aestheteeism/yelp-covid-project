```python
import json
import pandas as pd
```

```python
yelp_covid = pd.read_json("../data/yelp_academic_dataset_covid_features.json", lines=True)
yelp_covid = yelp_covid.drop_duplicates(subset=['business_id'], keep='first')
```

```python
yelp_covid
```

```python
# This is a different business dataset that was released alongside with the Yelp COVID19 dataset
yelp_business = pd.read_json("../data/yelp_academic_dataset_business_1.json", lines=True)
yelp_business
```

```python
yelp_merged = yelp_covid.merge(yelp_business[['business_id','name','state','stars','review_count','attributes','categories']], on='business_id')
yelp_merged
```

```python
# Get only fast-food related restaurant
only_fnb = yelp_merged.categories.str.contains('Restaurant', na=False) | yelp_merged.categories.str.contains('Food', na=False) | yelp_merged.categories.str.contains('Coffee', na=False)
yelp_merged = yelp_merged[only_fnb]
yelp_merged = yelp_merged.reset_index(drop=True)
yelp_merged
```

```python
yelp_merged.columns
```

```python
aDict = {"TRUE": True, "FALSE": False}

# Non-Exhaustive Mapping
cols = ['highlights', 'delivery or takeout', 'Grubhub enabled',
       'Call To Action enabled', 'Request a Quote Enabled', 'Covid Banner',
       'Temporary Closed Until', 'Virtual Services Offered']

def toBoolean(df, cols, di):
    for col in cols:
        df[col].map(di).fillna(df[col])
    return df 

yelp_merged = toBoolean(yelp_merged, cols, aDict)
yelp_merged
```
